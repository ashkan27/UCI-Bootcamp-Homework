```python
import pandas as pd
from sqlalchemy import create_engine
from config import user, password
```

### Extract

```python
csv_file = "nba.csv"
nba_player_data_df = pd.read_csv(csv_file)
nba_player_data_df.head()
```

```python
csv_file = "nba_extra.csv"
nba_extra = pd.read_csv(csv_file)
nba_extra.head()
```

```python
csv_file = "NBA_season1718_salary.csv"
NBA_season17_18_salary_data_df = pd.read_csv(csv_file)
NBA_season17_18_salary_data_df.head()
```

### Transformation

```python
player = nba_player_data_df["Player"]
player2 = nba_extra["Player"]
for i in range(len(player)):
    player[i] = player[i].partition("\\")[0]
    player2[i] = player2[i].partition("\\")[0]
```

```python
nba_player_data_df["Player"] = player
nba_extra["Player"] = player2

nba_extra.head()
```

```python
# To get he columns we need
nba_cols = ["Player", "Tm", "MP", "TS%", "PER", "ORB%", "DRB%"]
nba_transformed= nba_player_data_df[nba_cols].copy()

# Display
nba_transformed.head()
```

```python
# To get he columns we need
nba_cols = ["Player", "PTS", "AST", "BLK", "STL"]
nba_transformed2 = nba_extra[nba_cols].copy()

# Display
nba_transformed2.head()
```

```python
# Merge:

nba_stat = pd.merge(nba_transformed, nba_transformed2, on=["Player"])

# Rename and Sort
nba_stat = nba_stat.rename(columns={"Player": "player", "Tm": "tm", "MP": "mp", "TS%": "ts", "PER": "rating", "ORB%": "orb", "DRB%": "drb", "PTS": "pts", "AST": "ast", "BLK": "blk", "STL": "stl"})

# Delete players who have played less than 400 minutes in the season
nba_stat = nba_stat[nba_stat['mp'] > 400]  

nba_stat.set_index("player", inplace=True)

# Display
nba_stat.head()
```

```python
# NBA BIOOOO

nba_cols = ["Player", "Tm", "Pos","Age", "MP"]
nba_bio= nba_player_data_df[nba_cols].copy()

# Rename
nba_bio = nba_bio.rename(columns={"Player": "player", "Tm": "tm", "Pos": "pos", "Age": "age", "MP": "mp"})

# Delete players who have played less than 400 minutes in the season
nba_bio = nba_bio[nba_bio['mp'] > 400]

nba_bio.set_index("player", inplace=True)
nba_bio = nba_bio.drop(columns="mp")

# Display
nba_bio.head()
```

```python
# To get he columns we need
salary_cols = ["Player", "Tm", "seson17_18"]
salary_transformed= NBA_season17_18_salary_data_df[salary_cols].copy()

# Rename
nba_salary = salary_transformed.rename(columns={"seson17_18": "salary", "Player": "player", "Tm": "tm"})

nba_salary.set_index("player", inplace=True)

# Display
nba_salary.head()
```

#### Database Connection

```python
rds_connection_string = user + ":" + password + "@localhost:5432/nba_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

#### LOAD

```python
nba_stat.to_sql(name='nba_stat', con=engine, if_exists='append')
```

```python
nba_bio.to_sql(name='nba_bio', con=engine, if_exists='append')
```

```python
nba_salary.to_sql(name='nba_salary', con=engine, if_exists='append')
```
